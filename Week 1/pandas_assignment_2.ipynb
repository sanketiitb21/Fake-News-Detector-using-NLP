### Task 6. Which was the most ordered item

```python
# Note that just counting the number of times each item has occured in the table isn't
# enough because the quantity ordered isn't necessarily constant
import pandas as pd

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

chipotle_data = pd.read_csv(url, sep='\t')

# Group by item_name and sum the quantities
item_totals = chipotle_data.groupby('item_name')['quantity'].sum()

# Find the item with the highest total quantity
most_ordered_item = item_totals.idxmax()
max_quantity = item_totals.max()

print("Most ordered item:", most_ordered_item)
print("Total quantity ordered:", max_quantity)
```

### Task 11. What is the average revenue amount per order?

```python
import pandas as pd

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

chipotle_data = pd.read_csv(url, sep='\t')

# Convert item_price to numeric
chipotle_data['item_price'] = chipotle_data['item_price'].str.replace('$', '', regex=True).astype(float)

# Calculate revenue per row
chipotle_data['revenue'] = chipotle_data['quantity'] * chipotle_data['item_price']

# Total revenue per order
revenue_per_order = chipotle_data.groupby('order_id')['revenue'].sum()

# Average revenue per order
average_revenue = revenue_per_order.mean()
print("Average revenue per order: ${:.2f}".format(average_revenue))
```

### Task 15. What was the quantity of the most expensive item ordered?

```python
import pandas as pd

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

chipotle_data = pd.read_csv(url, sep='\t')

# Convert item_price to numeric
chipotle_data['item_price'] = chipotle_data['item_price'].str.replace('$', '', regex=True).astype(float)

# Find the maximum item price
max_price = chipotle_data['item_price'].max()

# Filter rows with the most expensive item
most_expensive_items = chipotle_data[chipotle_data['item_price'] == max_price]

# Total quantity of the most expensive item ordered
total_quantity = most_expensive_items['quantity'].sum()

print("Most expensive item price: $", max_price)
print("Quantity of the most expensive item ordered:", total_quantity)
```
